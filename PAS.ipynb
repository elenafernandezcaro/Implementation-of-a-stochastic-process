# Ejercicio: simular y visualizar PAS

```python
import numpy as np
from matplotlib import pyplot as plt
```

El PAS es un proceso estocástico definindo sobre $\mathbb{N}$.

Para cada $ t \in \mathbb{N} $, se considera $$ X_t = \sum_{s=1}^t Z_s$$ siendo $\{Z_s\}_{s\in \mathbb{N}}$ variables aleatorias independientes e identicamente distribuidas según la siguiente distribución:

$$ P(Z_s = 1) = p,\\ $$ $$ P(Z_s = -1 ) = 1-p. $$

Se le llama Paseo Aleatorio Simple de parámetro $p\in [0,1]$.

## Implementación de una función para simular el PAS

```python
from re import X
def simular_PAS(p,n,reps):
  # Input:
  # p: probabilidad de tomar el valor 1 en cada paso.
  # n: número de pasos a simular en cada trayectoria.
  # reps: número de trayectorias a simular.

  # Output: la función devuelve un array de numpy de dimensión (reps, n) que contiene los valores del PAS en cada paso para cada trayectoria.

  opciones=[1,-1]
  probabilidades=[p,1-p]
  pasos = np.random.choice(opciones, size=(reps, n), p=probabilidades)  # shape: (reps,n)
  output = np.cumsum(pasos, axis=1)

  return output  # shape: (reps, n)
```

Visualizamos una representación gráfica de 10 trayectorias diferentes de un PAS con 500 pasos y parámetro $p=0.5$.

```python
matriz_datos = simular_PAS(0.5,500,10)
plt.plot(np.transpose(matriz_datos))
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.ylabel("$X_t$")
plt.xlabel("$t$")
```

Para $p=0.5$:

```python
from matplotlib.lines import lineStyles
p = 0.5
n = 500
reps = 20
t = np.arange(n+1)

# Calculamos una matriz de datos con la función que simula las trayectorias.
matriz_datos = simular_PAS(p,n,reps)

# Calculamos la media y la desviación típica teóricas para esos p y n.
media = t*(2*p-1)
media_valor = (2*p-1)
desviacion_tipica = (np.sqrt(t*(4*p*(1-p))))
desviacion_tipica_valor=np.sqrt(4*p*(1-p))


# Calculamos los límites para la región del sombreado para ±1σ.
lim_sup_1 = media + desviacion_tipica
lim_inf_1 = media - desviacion_tipica

# Calculamos los límites para la región del sombreado para ±2σ.
lim_sup_2 = media + 2*desviacion_tipica
lim_inf_2 = media - 2*desviacion_tipica


# Graficamos las trayectorias del PAS añadiendo el sombreado de ±1σ y ±2σ.

plt.plot(np.transpose(matriz_datos), alpha = 0.7)  # Trayectorias.

plt.fill_between(t, lim_inf_1, lim_sup_1, color='gray', alpha=0.3)  # Sombreado.
plt.fill_between(t, lim_inf_2, lim_sup_2, color='gray', alpha=0.1)

plt.plot(t, media, color = 'black', linestyle = '--', linewidth = 1.5)  # Media teórica.

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)  # Cuadrícula.

plt.ylabel("$X_t$")
plt.xlabel("$t$")
```

## Visualización 3: visualización de la media empírica del PAS cuando aumentamos el número de trayectorias (junto con la media teórica)

Visualizamos media teórica junto con media empírica cuando
promediamos varias trayectorias.

```python
p = 0.5
n = 500

# Calulamos el vector de medias para la simulación del PAS con reps = 5, 100, 200, 500 y 1000.

media_5 = np.mean(simular_PAS(p,n,5), axis=0)       # shape : n*1
media_100 = np.mean(simular_PAS(p,n,100), axis=0)   # shape : n*1
media_200 = np.mean(simular_PAS(p,n,200), axis=0)   # shape : n*1
media_500 = np.mean(simular_PAS(p,n,500), axis=0)   # shape : n*1
media_1000 = np.mean(simular_PAS(p,n,1000), axis=0) # shape : n*1

# Calculamos la media teórica.

t = np.arange(n)
media = t*(2*p-1)

# Graficamos los vectores media (empíricas) junto con la media teórica.

matriz_medias = np.column_stack((media_5, media_100, media_200, media_500, media_1000))  #shape :n*5

nombres = ['5', '100', '200', '500', '1000']

plt.plot(matriz_medias, label=nombres, alpha = 0.8)   # Aquí no es necesario trasponer la matriz porque la función np.column_stack une los vectores por columnas.
plt.plot(media, color = 'black', linewidth = 1.8, label='Exacta')

plt.legend()
```

Se observa que cuantas más trayectorias
promediemos, mejor nos aproximaremos a la media teórica.
